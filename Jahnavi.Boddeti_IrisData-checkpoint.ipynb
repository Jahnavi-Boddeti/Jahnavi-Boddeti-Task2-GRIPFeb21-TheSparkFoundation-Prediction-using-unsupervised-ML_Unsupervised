**K-Means Clustering**

k-means clustering is one of the Unsupervised machine learning technique with the objective of putting the observations into k clusters in which each observation belongs to a cluster with the nearest mean.

The main input to the k-means clustering is the number of clusters (herein called k). k determines the clustering mechanism, and how the clusters form.The best way to find the number of clusters is by trial/error method.

once you know the number of clusters, there are three ways to assign the cluster centers:

* Manually
* Randomly
* "k-means++" in SKLearn

The latter selects initial cluster centers for k-mean clustering in a smart way to speed up convergence.

* *Assignment step:* Assign each observation to the cluster whose mean has the least squared Euclidean distance, this is intuitively the nearest mean.

* *update step:* Calculate the new means (centroids) of the observations in the new clusters.

* *Check for convergence:* The algorithm assumes convergenge when the assignments no longer change.








**Find the optimum number of clusters for k-means classification using The elbow method**


A fundamental step for any unsupervised algorithm is to determine the optimum number of clusters into which the data may be clustered. The Elbow method is one of the most popular methods to determine the optimum value of k.


```python
from  sklearn.cluster import KMeans
wcss = []
for i in range(1, 30):
    kmeans=KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 29, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
#plotting the results onto a line graph,
#allowing us to observe 'The elbow'
plt.plot(range(1, 30), wcss)
plt.title('The elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.grid()
plt.show()
```

From the above elbow graph we choose the optimum number of clusters as "3",this is where within cluster some of squares doesn't decrease significantly with every iteration. 

**Implementing KMeans Clustering**

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 29, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

**Data Visualization**

```python
# Visualizing the clusters on the first two columns
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'purple' , label = 'Iris-setosa' )
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'grey' , label = 'Iris- versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')
plt.legend()

```

```python
# 3d scatterplot using matplotlib
 
fig = plt.figure(figsize = (15, 15))
ax = fig.add_subplot(111, projection = '3d')
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'purple', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'grey', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'centroids')
plt.legend()
plt.show()
```

Hence by using The Elbow method the optimum number of clusters(k) are predicted to be ' 3 '
